# Normal Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
df = pd.read_csv('./data_pd/hs_heights_pair.csv')
df.head()
```

```python
boys_heights = df.boys
girls_heights = df.girls
```

```python
mean_boys = boys_heights.mean()
mean_girls = girls_heights.mean()

std_boys = boys_heights.std()
std_girls = girls_heights.std()

print(f"Boys: Mean = {mean_boys}, Standard Deviation = {std_boys}")
print(f"Girls: Mean = {mean_girls}, Standard Deviation = {std_girls}")
```

```python
# Define colors for the SD lines
colors = ['green', 'orange', 'red']

plt.figure(figsize=(14,6))
# Plot for boys
plt.subplot(1,2,1)
sns.histplot(boys_heights, kde=True, stat = 'density', linewidth = 0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_boys, std_boys)
plt.plot(x,p,'k',linewidth = 2)
title = "Boys' Heights\n Mean = {:.2f}, SD = {:.2f}".format(mean_boys, std_boys)
plt.title(title)

# Plot standard deviation lines for boys
for i in range(1, 4):
    plt.axvline(mean_boys + i * std_boys, color = colors[i-1], linestyle='--')
    plt.axvline(mean_boys - i * std_boys, color = colors[i-1], linestyle='--')
    
plt.axvline(mean_boys, color = 'yellow', linestyle='--')

# Plot for girls
plt.subplot(1,2,2)
sns.histplot(girls_heights, kde=True, stat = 'density', linewidth = 0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_girls, std_girls)
plt.plot(x,p,'k',linewidth = 2)
title = "Girls' Heights\n Mean = {:.2f}, SD = {:.2f}".format(mean_girls, std_girls)
plt.title(title)

# Plot standard deviation lines for girls
for i in range(1, 4):
    plt.axvline(mean_girls + i * std_girls, color = colors[i-1], linestyle='--')
    plt.axvline(mean_girls - i * std_girls, color = colors[i-1], linestyle='--')
    
plt.axvline(mean_girls, color = 'yellow', linestyle='--')
    
```

```python
df.dtypes
```

```python
df['networth'].isna().sum()
```

```python
# Calcultae z-scores for networth
df['networth_percentile'] = df['networth'].rank(pct=True) * 100
df.head()
```

```python
# Display mean and SD
networth_mean = df.networth.mean()
networth_std = df.networth.std()

print(f"Mean of networth: {networth_mean:.2f} B")
print(f"Standard deviation of networth: {networth_std:.2f} B")
```

# Shapiro Test

`shapiro test and QQ plot`

```python
from scipy.stats import shapiro
```

```python
from scipy import stats
```

```python
df = pd.read_csv('./data_pd/diabetes.csv')
df.head()
```

```python
# Plot a histogram of the Age column
plt.hist(df.Glucose, bins=30, edgecolor='k', alpha=0.7)
plt.title('Glucose Distribution')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate a QQ plot
stats.probplot(df.Glucose, dist='norm', plot=plt)
plt.title('Q-Q Plot of Glucose')
plt.show()
```

```python
# Perform the Shapiro-Wilk test
stat_shapiro, p_shapiro = shapiro(df.Glucose)
print('Shapiro-Wilk Test:')
print('Statistic=%.3f, p%.6f' % (stat_shapiro, p_shapiro))
```

```python
# Interpret the result
alpha = 0.5
if p_shapiro > alpha:
    print("Sample look Gaussion (fail to reject H0)")
else:
    print("Sample not look Gaussion (reject H0)")
    
    print("Sample look Gaussion (fail to reject H0)")
    print("Skewwness:", df.Glucose.skew())
```

# Central Limit Theorem

```python
sample_means = []
n_samples = 100
sample_size = 30

for _ in range(n_samples):
    sample = df['Glucose'].sample(n=sample_size, replace=True)
    sample_means.append(sample.mean())
```

```python
# Plot the distribution osample means
plt.hist(sample_means, bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Sample Means (n=30)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate QQ plot for sample mean
stats.probplot(sample_means, dist='norm', plot=plt)
plt.title('Q-Q Plot of Sample Mean')
plt.show()
```

```python
# Perform the Shapiro-Wilk test on  the sample means
stat_shapiro, p_shapiro = shapiro(sample_means)
print('Shapiro-Wilk Test:')
print('Statistic=%.3f, p%.6f' % (stat_shapiro, p_shapiro))
```

```python

# Interpret the result
alpha = 0.5
if p_shapiro > alpha:
    print("Sample look Gaussion (fail to reject H0)")
else:
    print("Sample not look Gaussion (reject H0)")
    
    print("Sample look Gaussion (fail to reject H0)")
    print("Skewwness:", sample_means.skew())
```
