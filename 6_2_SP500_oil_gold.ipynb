```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data_pd/SP500 oil gold bitcoin.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
cov = df[['BITCOIN', 'Gold', 'S&P500']].cov()
corr = df[['BITCOIN', 'Gold', 'S&P500']].corr()
```

```python
cov
```

```python
corr
```

```python
t1 = cov.loc['BITCOIN', 'Gold']
t2 = cov.loc['BITCOIN','S&P500']
t3 = corr.loc['BITCOIN', 'Gold']
t4 = corr.loc['BITCOIN','S&P500']
```

```python
t3
```

```python
# Create Scatter plot with correlation values
fig, axes = plt.subplots(2,2,figsize=(18,12))
# overall data plot
axes[0, 0].scatter(df['BITCOIN'],df['Gold'], color='blue', alpha=0.5)
# axes[0, 0].set_title('Overall Covariane : ', cov.loc['BITCOIN', 'Gold'])
axes[0, 0].set_title(f'Overall Covariane : {t1:.2f}')
axes[0, 0].set_xlabel('BITCOIN')
axes[0, 0].set_ylabel('Gold')
axes[0, 0].grid(True)

# Male Data plot
axes[0, 1].scatter(df['BITCOIN'],df['S&P500'], color='green', alpha=0.5)
# axes[0, 1].set_title('BITCOIN Covariane : ', cov.loc['BITCOIN','S&P500'])
axes[0, 1].set_title(f'BITCOIN Covariane : {t2:.2f}')
axes[0, 1].set_xlabel('BITCOIN')
axes[0, 1].set_ylabel('S&P500')
axes[0, 1].grid(True)

# overall data plot
axes[1, 0].scatter(df['BITCOIN'],df['Gold'], color='blue', alpha=0.5)
# axes[1, 0].set_title('Overall Correlation : ', corr.loc['BITCOIN', 'Gold'])
axes[1, 0].set_title(f'Overall Correlation : {t3:.2f}')
axes[1, 0].set_xlabel('BITCOIN')
axes[1, 0].set_ylabel('Gold')
axes[1, 0].grid(True)

# Male Data plot
axes[1, 1].scatter(df['BITCOIN'],df['S&P500'], color='green', alpha=0.5)
# axes[1, 1].set_title('BITCOIN Correlation : ', corr.loc['BITCOIN','S&P500'])
axes[1, 1].set_title(f'BITCOIN Correlation : {t4:.2f}')
axes[1, 1].set_xlabel('BITCOIN')
axes[1, 1].set_ylabel('S&P500')
axes[1, 1].grid(True)
```
