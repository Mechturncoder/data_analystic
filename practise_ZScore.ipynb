```python
ice_df = cream_df[['IPN31152N']].copy()
ice_df.head()
```

```python
ice_df['Diff12'] = ice_df['IPN31152N'].diff(6)    #this will show null to start with
ice_df.head(20)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
plot_pacf(ice_df['Diff12'].dropna(), lags=20)
plt.show()
```

```python
plot_acf(ice_df['Diff12'].dropna(), lags=20)
plt.show()
```

```python
train = ice_df[:round(len(ice_df)*66/100)]      # take the first 70% data

test = ice_df[round(len(ice_df)*66/100):]  
```

```python
# Now SARIMAX predictions
p,d,q = 3,1,2
model = SARIMAX(train['IPN31152N'], order=(p,d,q), seasonal_order=(1,0,2,12))   #(3,1,2)(1,0,2)[12] 
model_fit = model.fit()
prediction= model_fit.predict(start=test.index[0], end=test.index[-1])
ice_df['sarimaxPred'] = prediction
ice_df.tail()
```

```python
sns.lineplot(data=ice_df, x=ice_df.index, y='IPN31152N')

sns.lineplot(data=ice_df, x=ice_df.index, y='sarimaxPred')
plt.show()
```

```python
# grid search ARIMA parameters for time series
import warnings
from math import sqrt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# evaluate an ARIMA model for a given order (p,d,q)
def evaluate_arima_model(X, arima_order,order_seasonal):
	# prepare training dataset
	train_size = int(len(X) * 0.66)
	train, test = X[0:train_size], X[train_size:]
	history = [x for x in train]
	# make predictions
	predictions = list()
	for t in range(len(test)):
		model = SARIMAX(history, order=arima_order, seasonal_order=order_seasonal)
		model_fit = model.fit()
		yhat = model_fit.forecast()[0]
		predictions.append(yhat)
		history.append(test[t])
	# calculate out of sample error
	rmse = sqrt(mean_squared_error(test, predictions))
	return rmse

# evaluate combinations of p, d and q values for an ARIMA model
def evaluate_models(dataset, p_values, d_values, q_values):
    dataset = dataset.astype('float32')
    best_score, best_cfg = float("inf"), None
    t = 1
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                order_seasonal = (p,d,q,12)

                try:
                    print("Time:", t)
                    t+=1
                    rmse = evaluate_arima_model(dataset, order, order_seasonal)
                    print('ARIMA%s RMSE=%.3f' % (order,rmse))
                    if rmse < best_score:
                        best_score, best_cfg = rmse, order
                except:
                    continue
    print('Best ARIMA%s RMSE=%.3f' % (best_cfg, best_score))

# evaluate parameters
p_values = [1, 2]
d_values = range(1, 3)
q_values = range(1, 3)
warnings.filterwarnings("ignore")
evaluate_models(ice_df['IPN31152N'], p_values, d_values, q_values)
```
