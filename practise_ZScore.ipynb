```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data_pd/Men ODI Player Innings Stats - 21st Century.csv")
df.head()
```

```python
df_2016 = df[df['Innings Date'].str.contains('2016')]
df_2016['Innings Date']
```

```python
df_2016['Innings Date'] = pd.to_datetime(df_2016['Innings Date'].astype(str))
df_2016
```

```python
df_2016.isna().sum()
```

```python
df_2016 = df_2016.dropna(subset=['Innings Runs Scored'])
df_2016.isna().sum()
```

```python
df_2016['Innings Runs Scored'].unique()
```

```python
df_2016['Innings Runs Scored'] = df_2016['Innings Runs Scored'].str.extract('(\d+)')
df_2016['Innings Runs Scored'] = pd.to_numeric(df_2016['Innings Runs Scored'], errors='coerce')
df_2016 = df_2016.dropna(subset=['Innings Runs Scored'])
df_2016['Innings Runs Scored'].unique()
```

```python
df_2016 = df_2016.groupby(by='Innings Player').agg({'Innings Runs Scored': 'sum'})
df_2016.head(10)
```

```python
df_2016 = df_2016.sort_values(by='Innings Runs Scored', ascending=False)
df_2016.head(25)
```

```python
df_2016['Z_Scored'] = (df_2016['Innings Runs Scored'] - df_2016['Innings Runs Scored'].mean())/df_2016['Innings Runs Scored'].std()
df_2016.head(25)
```
